```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt='.1f', linewidths=.1)
plt.show()
```

```python
Sales_and_Profits = df.groupby("Segment").sum().iloc[:,[1,-1]].sum()
round(Sales_and_Profits,2)
```

```python
Top_5_Sales = df.groupby("State").Sales.sum().nlargest(n=5)
Top_5_Profits = df.groupby("State").Profit.sum().nlargest(n=5)
```

```python
Top_5_Sales.index
```

```python
Top_5_Profits.index
```

```python
Profit_sum = df.groupby("State")["Profit"].sum()
Profit_sum
```

```python
plt.figure (figsize=(16,12))
sns.barplot(x=Profit_sum.index, y=Profit_sum.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

```python
category_data = df.groupby(['Category'])[['Sales','Discount','Profit']].mean()
```

```python
category_data.plot.pie(subplots=True,figsize=(20,18),autopct='%0.1f%%',labels=category_data.index)
plt.show()
```

```python
shipmode_data = df.groupby(['Ship Mode'])[['Sales','Discount','Profit']].mean()
shipmode_data.plot.pie(subplots=True,figsize=(18,20),autopct='%0.1f%%',labels=shipmode_data.index)
plt.show()
```

# Step 4: relation between given attributes 
 


```python
plt.figure(figsize=(14,4))
sns.set(font_scale =1,palette = "viridis")
sns.barplot(data = df, x="Region",y="Profit",hue="Segment")
plt.show()
```

```python
plt.figure(figsize=(14,4))
sns.set(font_scale =1,palette = "viridis")
sns.barplot(data = df, x="Region",y="Profit",hue="Category")
plt.show()
```

```python
plt.figure(figsize=(20,4))
sns.set(font_scale =1,palette = "viridis")
sns.barplot(data = df, x="State",y="Profit",hue="Sub-Category")
plt.title("Investigation of Central Region Furniture Category: profit analysis  ")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True,bw=25)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

```python
x= df.groupby('State')[['Sales','Profit']].sum().sort_values(by="Sales",ascending=False)
x[:].plot.bar(color=['red','green'],figsize=(15,8))
plt.title('Profit/loss & sales across states' )
plt.xlabel('States')
plt.ylabel("Profit/loss & sales")
plt.show()
```
